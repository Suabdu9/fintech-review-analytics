# FinTech Review Analytics: Customer Sentiment Analysis of Ethiopian Banking Apps

## 1. Introduction

### Digital Banking Growth
Digital banking has rapidly expanded in recent years, fueled by advances in technology, widespread smartphone adoption, and customer demand for convenient financial services. Traditional banks are modernizing, while fintechs are introducing innovative solutions that make banking more accessible and efficient.

### Importance of Customer Reviews
Customer reviews play a crucial role in shaping perceptions of digital banking platforms. Positive feedback builds trust and attracts new users, while negative reviews highlight areas for improvement. Reviews provide valuable insights into customer satisfaction, usability, and reliability.

### Project Objectives
This project aims to analyze customer reviews of digital banking platforms to identify key trends, sentiments, and improvement opportunities. Specifically, the objectives are to:
- Understand customer expectations and pain points  
- Provide actionable recommendations for enhancing user experience  
- Support the growth and competitiveness of digital banking services  


## 2. Dataset Overview

```python
import pandas as pd

df = pd.read_csv(
    "../data/processed/analyzed_bank_reviews.csv"
)

df.head()
```

```python
df["bank"].value_counts()
```

## 3. Sentiment Distribution by Bank

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.countplot(
    data=df,
    x="bank",
    hue="sentiment_label"
)

plt.title("Sentiment Distribution by Bank")

plt.xlabel("Bank")

plt.ylabel("Number of Reviews")

plt.xticks(rotation=10)

plt.tight_layout()

plt.show()
```

```python
plt.savefig(
    "../plots/sentiment_distribution.png"
)
```

### Insight

Commercial Bank of Ethiopia received the highest volume of reviews, indicating strong user engagement and adoption. Across all banks, positive reviews dominate overall sentiment; however, negative reviews remain significant and are largely associated with authentication failures, transaction delays, and app stability concerns.

## 4. Rating Distribution per Bank

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    data=df,
    x="bank",
    y="rating"
)

plt.title("Rating Distribution per Bank")

plt.xlabel("Bank")

plt.ylabel("Rating")

plt.tight_layout()

plt.show()
```

```python
plt.savefig(
    "../plots/rating_distribution.png"
)
```

### Insight

Higher ratings are strongly associated with positive sentiment labels, while lower ratings frequently correspond to complaints related to OTP verification failures, slow transaction processing, and application crashes.

## 5. Top Themes and Keywords

```python
theme_counts = (
    df["identified_theme"]
    .value_counts()
    .reset_index()
)

theme_counts.columns = [
    "Theme",
    "Count"
]

plt.figure(figsize=(10, 6))

sns.barplot(
    data=theme_counts,
    x="Count",
    y="Theme"
)

plt.title("Most Frequent Themes")

plt.tight_layout()

plt.show()
```

```python
plt.savefig(
    "../plots/theme_frequency.png"
)
```

### Insight

Thematic analysis revealed that account access issues and transaction performance problems are the most frequent customer concerns across all banking applications. Positive themes primarily focused on convenience, accessibility, and ease of use.

## 6. Bank-Level Insights

### Commercial Bank of Ethiopia (CBE)

Satisfaction Drivers:
- Convenient mobile transaction services
- Broad user adoption and accessibility

Pain Points:
- OTP and login authentication failures
- Slow transaction confirmation

Recommendations:
- Improve authentication reliability
- Optimize transaction response time infrastructure

### Bank of Abyssinia (BOA)

Satisfaction Drivers:
- Positive user interface feedback
- Easy navigation experience

Pain Points:
- App crashes and freezing
- Delayed loading performance

Recommendations:
- Improve application stability testing
- Optimize app performance for low-end devices

### Dashen Bank

Satisfaction Drivers:
- Feature-rich mobile banking experience
- Positive customer convenience feedback

Pain Points:
- Login session issues
- Feature request complaints

Recommendations:
- Improve session management reliability
- Introduce customer-requested usability features

## 7. Comparative Analysis

Positive sentiment dominates across all three banking applications, indicating generally favorable customer perception of Ethiopian digital banking services.

However, recurring operational issues appear consistently across banks, particularly:
- authentication reliability
- transaction delays
- app stability

Commercial Bank of Ethiopia generated the highest engagement volume, while Bank of Abyssinia showed relatively stronger UI-related feedback. Dashen Bank reviews reflected stronger feature-related discussions and customer enhancement requests.
